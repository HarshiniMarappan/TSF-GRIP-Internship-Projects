### Statistical summary

```python
data[['Sales','Quantity','Discount','Profit']].describe()
```

### Finding uniques entries and most occuring entries in a column

```python
cols = data.columns.values.tolist()
cols
```

```python
for col in cols[0:5] + cols[6:9]:
    print(col,'\t','Unique Values : ',data[col].unique())
    print('\t\t','Most Occurring : ',st.mode(data[col]))
    print()
```

## Conclusion drawn from above analysis

**There seems to be high fluctuation in profit and sales as seen from the statistical summary .**

**Transportation cost :** Has been kept as minimum as possible , standard class is the least expensive mode

**Segment :** Has sold more of domestic products rather than commercial

**Category :** More number of office products has been sold

**Subcategory :** Most sold is Binders

**Maximum sales location :** City wise its Newyork , state wise its California, region wise its Western US
